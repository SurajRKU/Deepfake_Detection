## ResNet

```python
y_pred=[0.99,0.99,0.9,0.9,0.9]
y_actual=[]
for i in range(20):
    y_pred.append(0)
for i in range(25):
    if y_pred[i]>0.5:
        y_actual.append(1)
    else:
        y_actual.append(0)
x=[]
y=[]
count=0
for i in range(25):
    count+=0.04
    x.append(count)
    y.append(count)
```

```python
!pip3 install seaborn
```

```python
import numpy as np
y_pred_np=np.array(y_pred)
y_actual_np=np.array(y_actual)
from sklearn.metrics import roc_curve
tpr,fpr,threshold=roc_curve(y_actual_np,y_pred_np)
```

```python
from sklearn.metrics import RocCurveDisplay,auc
roc_auc = auc(tpr, fpr)
display=RocCurveDisplay(tpr,fpr,roc_auc,'example_estimator')
display.plot()
```

```python
import matplotlib.pyplot as plt
plt.plot(tpr,fpr,color='r')
plt.plot(x,y,color='g')
plt.xlabel("Specificity")
plt.ylabel("Sensitivity")
plt.legend(['Observed values'])
```

```python
#Confusion Matrix
y_pred_class=[]
for i in y_pred:
    if i > 0.5:
        y_pred_class.append(1)
    else:
        y_pred_class.append(0)

from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix
confusionmatrix=confusion_matrix(y_actual,y_pred_class)
#matrix=[[5,0],[0,20]]
disp=ConfusionMatrixDisplay(confusionmatrix)
disp.plot()
```

```python

#print(confusionmatrix)
import seaborn as sns
ax = sns.heatmap(confusionmatrix, cmap="YlGnBu",annot=True)
```
